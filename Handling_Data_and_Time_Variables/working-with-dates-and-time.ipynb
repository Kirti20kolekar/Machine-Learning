# Working with Dates

```python
# Converting to datetime datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

#### Extract year

```python
date['date_year'] = date['date'].dt.year

date.sample(5)
```

#### Extract Month

```python
date['date_month_no'] = date['date'].dt.month

date.head()
```

```python
date['date_month_name'] = date['date'].dt.month_name()

date.head()
```

#### Extract Days

```python
date['date_day'] = date['date'].dt.day

date.head()
```

```python
# day of week
date['date_dow'] = date['date'].dt.dayofweek

date.head()
```

```python
# day of week - name

date['date_dow_name'] = date['date'].dt.day_name()

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
# is weekend?

date['date_is_weekend'] = np.where(date['date_dow_name'].isin(['Sunday', 'Saturday']), 1,0)

date.drop(columns=['product_id','city_id','orders']).head()
```

#### Extract week of the year

```python
# Assuming 'date' column is a datetime type
date['date_week'] = date['date'].dt.isocalendar().week

# Dropping unnecessary columns
date = date.drop(columns=['product_id', 'city_id', 'orders'])

# Display the first few rows
date.head()
```

#### Extract Quarter

```python
# Extracting the quarter from the 'date' column
date['quarter'] = date['date'].dt.quarter

# Dropping unnecessary columns (modifies the DataFrame in place)
#date.drop(columns=['product_id', 'city_id', 'orders'], inplace=True)

# Displaying the first few rows
date.head()
```

#### Extract Semester

```python
date['semester'] = np.where(date['quarter'].isin([1,2]), 1, 2)

#date.drop(columns=['product_id','city_id','orders']).head()
date.head()
```

```python
# Months passed

import numpy as np
import pandas as pd

# Assuming 'today' is the current date
today = pd.Timestamp.today()

# Calculate the number of months passed using a more accurate method with pandas
date['months_difference'] = (today.year - date['date'].dt.year) * 12 + (today.month - date['date'].dt.month)

# Displaying the updated DataFrame
date.head()
```
